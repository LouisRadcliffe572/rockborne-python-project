# Noughts and Crosses

```python
def game():

    ## === Preamble ===
    
    # = Initial variable definitions =  
    names = []                 # player names are put into this list
    scores = [0,0]             # player scores are put into this list
    p_chars = ['','']          # player characters are put into this list
    p_chars_choice = ['','']   # player characters (verbal, for display)
    game = 1                   # game number (starts at 1)
    exit = ''                  # variable that ends the function
    
    # = Input player names =    
    names.append(input("Input a player name:"))
    names.append(input("Input another player name:"))
    
    # = Randomly decide which player starts =    
    import random
    starter = random.randint(0,1)   # randomly decides who plays first
    fixed = starter                # makes a copy of this variable that won't change later
    print(f'{names[fixed]} will go first!')
    
    # = Loop to ensure character choice is valid ('o' or 'x') =    
    while  p_chars[fixed] != 'o' and p_chars[fixed] != 'x':      # checks for valid character input
        p_chars[fixed] = input(f'{names[fixed]}, noughts or crosses? (o/x)').lower()
        if p_chars[fixed] != 'o' and p_chars[fixed] != 'x':      # checks for valid character input to prompt and ask again
            print("You didn't input 'o' or 'x'")
  
    # = Assigns characters to players =    
    if p_chars[fixed] == 'o':               # first player chooses noughts
        p_chars[1-fixed] = 'x'                # second player is assigned crosses
        p_chars_choice[fixed] = 'noughts'
        p_chars_choice[1-fixed] = 'crosses'
    else:                            # first player chooses crosses
        p_chars[1-fixed] = 'o'                # second player is assigned noughts
        p_chars_choice[fixed] = 'crosses'
        p_chars_choice[1-fixed] = 'noughts'
    
    print(f'{names[fixed]} is {p_chars_choice[fixed]}, {names[1-fixed]} is {p_chars_choice[1-fixed]}.')   # displays which player is which

    
    ## === Tournament loop - (potentially) multiple games ===    
    while True:   # ends when user inputs 'exit' after a game

        # = Further variable definitions =        
        board = [['_', '_', '_'], ['_', '_', '_'], ['_', '_', '_']]   # this is the game board (starts empty)
        moves = [[0, 0, 0], [0, 0, 0], [0, 0, 0]]                     # this shows whether a position has been played in (starts empty)
        turn = 1                                                      # turn number (starts as 1)
        result = ''                                                   # result of the current game (starts as null)

        # = Decides who starts the game =
        if game%2 == 1:
            starter = fixed
        else:
            starter = 1-fixed
        
        # = Print a "__'s turn to start" message is not the first game =
        if game > 1:
            if names[starter][len(names[starter])-1].lower() == 's':
                print(f"It is {names[starter]}' turn to start.")
            else:
                print(f"It is {names[starter]}'s turn to start.")
        
        # = Display empty game board as a grid ahead of game =       
        for row in range(0,3):             # 'row' and 'column' of 'board' variable
            for column in range(0,3):
                print(board[row][column], end='')
                if column < 2:
                    print(' | ', end='')
                else:
                    print()


        ## == Game loop - multiple turns ==        
        while result == '':   # while loop for current game, ends when a result is decided (player win or draw)
            
            # = Decides whose turn it is =
            if turn%2 == 1:                 # turn number is odd - first player's turn
                player_turn = starter
            else:                           # turn number is even - second player's turn
                player_turn = 1 - starter   # (if decider=1 then 1-decider=0 and vice versa, so each player index is covered)
            
            # = Player inputs where they would like to play =
            p_input = input(f'{names[player_turn]}, where would you like to play?')
                # location input as 'xy' where:
                #     x = t (top),  m (middle), or b (bottom)
                #     y = l (left), m (middle), or r (right) 
                # e.g. tm = top middle, mr = middle right

            # = Checks input is recognised =
            if p_input not in ['tl','tm','tr','ml','mm','mr','bl','bm','br']:   # checks against all valid location inputs
                print("I don't recognise that input!")
                turn -= 1                 # cancels out the turn increase increment at the end of the current game while loop and repeats
            else:
                if   p_input[0] == 't':   # row character
                    i = 0                 # top row
                elif p_input[0] == 'm':
                    i = 1                 # middle row
                elif p_input[0] == 'b':
                    i = 2                 # bottom row

                if   p_input[1] == 'l':   # column character
                    j = 0                 # left column
                elif p_input[1] == 'm':
                    j = 1                 # middle column
                elif p_input[1] == 'r':
                    j = 2                 # right column
                
                # = Checks that space is available & updates the board =
                if moves[i][j] == 1:      # checks if the chosen position is available to play in
                    print('That place is already taken!')
                    turn -= 1             # cancels out the turn increase increment at the end of the current game while loop and repeats
                else:
                    moves[i][j] = 1             # sets the chosen position in 'moves' to 1 (meaning that position has now been played in)
                    board[i][j] = p_chars[player_turn]   # updates chosen position on board with character of player whose turn it is

            # = Displays updated board =
            for row in range(0,3):              # displays updated game board
                for column in range(0,3):
                    print(board[row][column], end='')
                    if column < 2:
                        print(' | ', end='')
                    else:
                        print()

            ## == Win conditions - mutually exclusive ==
            
            # = Row win =
            if board[i][0] == board[i][1] and board[i][0] == board[i][2]:   # cells are equal across the row just played in
                result = f'{names[player_turn]} wins!'                      # win by row
                scores[player_turn] += 1                                    # gives the winning player a point
 
            # = Column win =
            elif board[0][j] == board[1][j] and board[0][j] == board[2][j]:   # cells are equal across the column just played in
                result = f'{names[player_turn]} wins!'                        # win by column
                scores[player_turn] += 1                                      # gives the winning player a point
            
            # = Diagonal win =
            elif moves[1][1] == 1 and ((moves[0][0] == 1 and moves[2][2]) or (moves[0][2] == 1 and moves[2][0] == 1)):   # checks first if either diagonal is full
                if   board[1][1] == board[0][0] and board[1][1] == board[2][2]:   # cells are equal across TopLeft to BottomRight diagonal
                    result = f'{names[player_turn]} wins!'                        # win by diagonal
                    scores[player_turn] += 1                                      # gives the winning player a point
                elif board[1][1] == board[2][0] and board[1][1] == board[0][2]:   # cells are equal across TopRight to BottomLeft diagonal
                    result = f'{names[player_turn]} wins!'                        # win by diagonal
                    scores[player_turn] += 1                                      # gives the winning player a point

                # = Draw (board is full, no win conditions met) =
                elif moves == [[1, 1, 1], [1, 1, 1], [1, 1, 1]]:    # if the game board is entirely full the game ends
                    result = "Draw! You'll have to play again!"     # no win condition has been met, so the game is a draw (no points added)

            turn += 1   # increments turn number to decide whose turn, repeats while loop

        # = Game results - only occurs once a 'result' has been reached (win or draw)
        print(result)                                                                            # prints result of this game
        print('Scores:')
        print(f'{names[fixed]}: {scores[fixed]}, {names[1-fixed]}: {scores[1-fixed]}')   # prints running total score
        
        # = Option to end tournament (and function) =
        exit = input("Keep playing? (type 'exit' to stop playing)")   # this is the only way to end the function

        # = Tournament results - only occurs once player chooses to end
        if exit.lower() == 'exit':
            print('Final Scores:')
            print(f'{names[fixed]}: {scores[fixed]}, {names[1-fixed]}: {scores[1-fixed]}, Total games = {game}')   # prints final score total
            
            # = Determines overall winner =
            if scores[fixed] > scores[1-fixed]:    # compares total scores to decide overall winner
                winner = names[fixed]
                print(f'{winner} is the winner!')
            elif scores[fixed] < scores[1-fixed]:
                winner = names[1-fixed]
                print(f'{winner} is the winner!')
            else:
                print('It was a draw!')
            print('Thank you for playing!')
            
            break                                  # ends the function
        else:
            #starter = 1 - starter   # this makes the starting player alternate across a tournament
            game += 1               # increments number of games
```

```python
game()
```
